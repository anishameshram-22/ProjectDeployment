```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv")
```

```python
import matplotlib.pyplot as plt


x = df['Experience Years']
y = df['Salary']
plt.title('Salary vs. Experience Years')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.scatter(x,y)

plt.show()
```

```python
df.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=2529)
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[2.5]])
```

```python
lr.predict([[10]])
```

```python
year=int(input("enter your experience in year:"))
salary=lr.predict([[year]])
print("your salary is:",salary)
```

```python
x = df['Experience Years']
y = df['Salary']
plt.figure(figsize=(10, 3))
plt.scatter(x,y)
plt.plot(x, lr.predict(x.values.reshape(-1, 1)), color='red')
plt.title('Salary vs. Experience Years')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.scatter(x,y)

plt.show()
```
